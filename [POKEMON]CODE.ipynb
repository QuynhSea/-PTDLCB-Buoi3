# **CLEAR**

```python
#Kiểm tra các giá trị Null hoặc NaN có trong dữ liệu
df_pokemon.isnull().sum()
```

```python
#xóa 2 cột không cần thiết:pokedex_number, percentage_male
if 'pokedex_number' in df_pokemon.columns and 'percentage_male' in df_pokemon.columns:
  df_pokemon.drop(columns=['pokedex_number', 'percentage_male'], axis=1, inplace=True)
  print(df_pokemon)
else:
  print("Columns 'pokedex_number' and/or 'percentage_male' not found in the DataFrame.")
```

```python
# Thay thế các giá trị trong type2 bằng None
df_pokemon['type2'].fillna('None', inplace=True)

# Thay thế các giá trị trong height_m bằng Mode của tất cả các cột
df_pokemon['height_m'].fillna(df_pokemon['height_m'].mean(), inplace=True)

# Thay thế giá trị trong weight_kg bằng Mode của tất cả các cột
df_pokemon['weight_kg'].fillna(df_pokemon['weight_kg'].mean(), inplace=True)
```

```python
df_pokemon.sample(5).T
```

```python
# Thêm total abilities mà pokemon có
df_pokemon["tot_abilities"] = df_pokemon.apply(lambda x: len(x["abilities"]), axis=1)
```

```python
# hợp nhất type1 và type2 thành cột mới type
# đổi tên type 1 thành primary và type 2 thành secondary
df_pokemon['type'] =  df_pokemon.apply(lambda x: x['type1'] if pd.isnull(x['type2']) else f'{x["type1"]}_{x["type2"]}', axis=1)
df_pokemon.rename(columns = {'type1':'primary type', 'type2':'secondary type'}, inplace = True)
```

2. Các yếu tố cơ bản ảnh hưởng thế nào đến Tốc độ của Pokemon ?

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 2, figsize=(16, 10), sharey=True)
sns.scatterplot(x=df_pokemon['attack'], y=df_pokemon['speed'], ax=axes[0, 0])
axes[0, 0].set_title("Speed V/S Attack")

sns.scatterplot(x=df_pokemon['defense'], y=df_pokemon['speed'], ax=axes[0, 1])
axes[0, 1].set_title("Speed V/S Defence")

sns.scatterplot(x=df_pokemon['height_m'], y=df_pokemon['speed'], ax=axes[1, 0])
axes[1, 0].set_title("Speed V/S Height")

sns.scatterplot(x=df_pokemon['weight_kg'], y=df_pokemon['speed'], ax=axes[1, 1])
axes[1, 1].set_title("Speed V/S Weight")

fig.suptitle("Speed Factor?", size=20)
plt.tight_layout()
plt.show()
```

3. Hệ Pokemon nào dễ bắt nhất?

```python
plt.figure(figsize=(16, 6))
ax = sns.boxplot(x='primary type', y='capture_rate', hue='is_legendary', data=df_pokemon)

ax.set_xlabel(xlabel='Primary Type')
ax.set_ylabel(ylabel='Capture Rate')
ax.set_title('Pokémon Capture Rate by Primary Type', pad=40)

sns.despine(top=True, right=True)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ['Non-legendary', 'Legendary'], loc=(1.1, 1.1))  # Adjusted legend position
plt.show()
```

4. Các hệ pokemon phổ biến nhất ở cả
hệ chính và hệ phụ là gì?

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 9))

# Primary (Top 5)
ax = sns.countplot(
    x='primary type',
    data=df_pokemon,
    order=df_pokemon['primary type'].value_counts().iloc[:5].index,
    ax=axes[0, 0]
)
ax.set_title('Primary Type')
for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x() + 0.25, p.get_height() + 0.01))

# Secondary (Top 5)
ax = sns.countplot(
    x='secondary type',
    data=df_pokemon,
    order=df_pokemon['secondary type'].value_counts().iloc[:5].index,
    ax=axes[0, 1]
)
ax.set_title('Secondary Type')
for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x() + 0.25, p.get_height() + 0.01))

# Primary (Least Frequent 5)
ax = sns.countplot(
    x='primary type',
    data=df_pokemon,  # Pass the DataFrame explicitly
    order=df_pokemon['primary type'].value_counts(ascending=True).index[:5],
    ax=axes[1, 0]
)
ax.set_title('Primary Type (Least Frequent)')
for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x() + 0.25, p.get_height() + 0.01))

# Secondary (Least Frequent 5)
ax = sns.countplot(
    x='secondary type',
    data=df_pokemon,  # Pass the DataFrame explicitly
    order=df_pokemon['secondary type'].value_counts(ascending=True).index[:5],
    ax=axes[1, 1]
)
ax.set_title('Secondary Type (Least Frequent)')
for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x() + 0.25, p.get_height() + 0.01))

plt.show()
```

5. Pokemon hệ nào mạnh nhất?

```python
top10_pokemon_base_total = df_pokemon.sort_values(by="base_total", ascending=False).reset_index()[:10]
plt.figure(figsize=(20,10))
ax = sns.barplot(x=top10_pokemon_base_total["NAME"], y=top10_pokemon_base_total["base_total"], orient='v')
ax.set_title("Which is the best pokémon?", size=20)
ax.set(xlabel="Base Total", ylabel="Name")
for p in ax.patches:
    ax.annotate('{:.1f}'.format( p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
```

6. Pokemon mạnh nhất theo chỉ số V/S Pokemon mạnh nhất theo cốt truyện

```python
attributes=['attack', 'sp_attack', 'defense', 'sp_defense', 'hp', 'speed']
# Mewtwo, Rayquaza
mewtwo= top10_pokemon_base_total[top10_pokemon_base_total['NAME'] == 'Mewtwo'][attributes].values.tolist()[0]
rayquaza = top10_pokemon_base_total[top10_pokemon_base_total['NAME'] == 'Rayquaza'][attributes].values.tolist()[0]
# Arceus
Arceus = top10_pokemon_base_total[top10_pokemon_base_total['NAME'] == 'Arceus'][attributes].values.tolist()[0]

angles=np.linspace(0,2*np.pi,len(attributes), endpoint=False)
angles=np.concatenate((angles,[angles[0]]))

attributes.append(attributes[0])
mewtwo.append(mewtwo[0])
rayquaza.append(rayquaza[0])
Arceus.append(Arceus[0])
```

```python
fig=plt.figure(figsize=(20,10))
ax=fig.add_subplot(111, polar=True)

#Arceus Plot
ax.plot(angles,Arceus, 'o-', color='blue', linewidth=1, label='Rayquaza')
ax.fill(angles, Arceus, alpha=0.25, color='skyblue')

# Mewtwo Plot
ax.plot(angles, mewtwo, 'o-', color='darkred', linewidth=1, label='Mewtwo')
ax.fill(angles, mewtwo, alpha=0.25, color='darkred')

# Rayquaza Plot
ax.plot(angles,rayquaza, 'o-', color='indianred', linewidth=1, label='Arceus')
ax.fill(angles, rayquaza, alpha=0.25, color='indianred')

ax.set_thetagrids(angles[:-1] * 180/np.pi, attributes[:-1], fontsize=12)
plt.grid(True)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ['Arceus', 'Mewtwo', 'Rayquaza'], loc=(0,0.99))

ax.set_title("Mewtwo & Rayquaza V/S Arceus", pad=40)
```

# **MA TRẬN TƯƠNG QUAN**

```python
#chuyển dữ liệu từ object sag float
def try_convert_to_float(value):
  try:
    return float(value)
  except:
    return np.nan

df_pokemon['NAME'] = df_pokemon['NAME'].apply(try_convert_to_float)
df_pokemon['abilities'] = df_pokemon['abilities'].apply(try_convert_to_float)
df_pokemon['classfication'] = df_pokemon['classfication'].apply(try_convert_to_float)
df_pokemon['primary type'] = df_pokemon['primary type'].apply(try_convert_to_float)
df_pokemon['secondary type'] = df_pokemon['secondary type'].apply(try_convert_to_float)
df_pokemon['type'] = df_pokemon['type'].apply(try_convert_to_float)
```

```python
against=[]
rest=[]
for i in df_pokemon.columns:
    if 'against' in i:
        against.append(i)
    else:
        rest.append(i)
```

```python
#Trực quan các chỉ số tấn công
fig,axes = plt.subplots(figsize=(20,10))
sns.heatmap(df_pokemon[against].corr(),annot=True).set_title('Chi so tan cong')
plt.show()
```
